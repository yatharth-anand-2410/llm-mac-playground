Even though the training crashed, it successfully saved a snapshot at iteration 100! 

This is the magic of **LoRA** (Low-Rank Adaptation). The checkpoint is just a tiny file containing the "changes" we want to make to the model. 

To run it, we load the giant 4-bit Llama 3 model into memory, and then we load our tiny `0000100_adapters.safetensors` file on top of it, mathematically patching the new behaviors into the brain of the AI.

```python
from mlx_lm import load, generate

# 1. Define the base model and our new adapter path
model_id = "mlx-community/Meta-Llama-3-8B-Instruct-4bit"

# We point the adapter specifically to the Iteration 100 checkpoint file
adapter_path = "adapters/"

print("Loading Base Model + Custom Adapter...")
# The load function is smart enough to layer the adapter file directly over the model
model, tokenizer = load(model_id, adapter_path=adapter_path)
print("\nâœ… Custom Model Loaded Successfully!")
```

```python
# 2. Test out the fine-tuned model
messages = [
    {"role": "system", "content": "You are a helpful, smart AI assistant. Answer the user's questions clearly & concisely."},
    {"role": "user", "content": "What is the meaning of life?"}
]

prompt = tokenizer.apply_chat_template(
    messages, 
    tokenize=False, 
    add_generation_prompt=True
)

print("Generating custom response...\n")
response = generate(
    model,
    tokenizer,
    prompt=prompt,
    verbose=True,      
    max_tokens=200
)
```
